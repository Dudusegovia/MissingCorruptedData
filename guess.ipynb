# Banco de dados completo enumerado.

```python
import pandas as pd
import numpy as np

data = {
    'BIAS': [1 for i in range(30)],
    "X1": [3, 4, 7, 1, 6, 1, 7, 1, 7, 8, 7, 8, 6, 9, 5, 9, 6, 9, 6,1,1,7,2,4,6,6,7,5,3,1],
    "X2": [1, 5, 5, 7, 5, 5, 2, 3, 9, 2, 7, 3, 6, 5, 8, 5, 3, 6, 7,6,1,6,3,2,1,9,7,2,6,9],
    "X3": [7, 2, 5, 5, 2, 1, 2, 9, 6, 3,2,2,3,4,4,1,6,7,3,7,9,3,5,7,4,5,7,8,6,2],
    "X4": [5, 2, 1, 8, 2, 6, 9, 4,3,1,2,9,9,3,9,9,5,1,4,4,7,5,3,4,9,6,5,8,1,5],
    "Y":  [4, 5, 5, 3, 5, 4, 3, 4,4,5,5,3,3,4,3,3,4,5,4,4,4,4,4,4,3,4,4,3,5,4]
       }
df = pd.DataFrame(data)
print(df)
```

```python
y = df['Y']
y
```

```python
x = df.drop(columns='Y')
x
```

# Aplicação da pseudoinversa na matriz de dados

# X^t



```python
xt = x.T
xt
```

# X^t * X


```python
xtdf = xt.dot(x)
xtdf
```

# (X^t * X)^-1


```python
inv = np.linalg.inv(xtdf)
inv
```

# (X^t * X)^-1 * X^t


```python
invdot = inv.dot(xt)
invdot
```

# (X^t * X)^-1 * X^t * Y

```python
global bias
bias = invdot.dot(y)
bias
```

# Aplicando e obtendo os falores de F para cada linha (exemplo)

```python
import copy as cp
def func(row):
    global bias
    return bias[0] + row["X1"]*bias[1] + row["X2"]*bias[2] + row["X3"]*bias[3] + row["X4"]*bias[4]
    
df_novo = cp.deepcopy(df)
df_novo['F'] = df.apply(func,axis=1)
df_novo
```

#### Definindo funções para achar as B e o Y

```python
import copy as cp
def acha_bias(df):
    y = df['Y']
    x = df.drop(columns='Y')

    xt = x.T
    bias = ((np.linalg.inv(xt.dot(x))).dot(xt)).dot(y)
    return bias

def func(row):
    global bias
    return bias[0] + row["X1"]*bias[1] + row["X2"]*bias[2] + row["X3"]*bias[3] + row["X4"]*bias[4]
```
